```python
weekH=sh1.col_values(11) # Assigning the column 52 Week High in variable weekH
weekH
```

```python
price=sh1.col_values(8) # Assigning the column price High in variable price
price
```

```python
len(price)              #checking total number of rows in price column
```

```python
for i in range(1,len(weekH)):     #Removing the comma in values of weekH and then converting it to float
    weekH[i]=float(weekH[i].replace(',',''))
```

```python
for i in range(1,len(price)):     #Removing the comma in values of price and then converting it to float
    price[i]=float(price[i].replace(',',''))
```

**Calculating Delta by ((52 week high - price) / 52 week high)**

```python
Delta=[]  
temp=0
for i in range(1,len(weekH)):
    temp=((weekH[i])-(price[i]))/(weekH[i])
    Delta.append(temp)
```

```python
Delta     #
```

**Assigning the cell AP1 as Delta on google sheet 1**

```python
sh1.update('AP1','Delta')
```

**Exporting the values of Delta on google sheet 1**

```python
cell_list = sh1.range('AP2:AP')
for i, val in enumerate(Delta):  
    cell_list[i].value = val    
    
sh1.update_cells(cell_list)
```

**Importing pandas and creating a series**

```python
import pandas as pd
ser1 = pd.Series(company)
ser2 = pd.Series(marCap)
ser3 = pd.Series(ten_year_Ret)
ser4 = pd.Series(dividend)
```

**Merging all the series in df using concat function**

```python
df = pd.concat([ser1, ser2,ser3,ser4], axis=1).reset_index()
df.columns = df.iloc[0]
df
```

```python
df = df[1:]            #removing row 0
df = df.drop(0,axis=1) #dropinh index
df = df.fillna(0)      #filling null values    
df
```

**Sorting the values of Divident in descending**

```python
df = df.sort_values('Dividend Per Share',ascending=False)
df
```

**Calculating High risk**

```python
highRisk = df[(df['Market Cap(Cr)'] < 4000) & (df['10-Year Return(%)'] < 8)]
highRisk
```

**Assigning Top 5 values in highRisk**

```python
highRisk=highRisk.head(5)
highRisk
```

**Calculating Risk**

```python
risk = df[(df['Market Cap(Cr)'] >= 4000) & (df['Market Cap(Cr)']<7000) & 
        (df['10-Year Return(%)'] >= 8) & (df['10-Year Return(%)']<15)]
risk = risk.head(5)
risk
```

**Calculating Moderate Risk**

```python
modRisk = df[(df['Market Cap(Cr)'] >= 7000) & (df['Market Cap(Cr)'] < 17000) & 
           (df['10-Year Return(%)'] >= 15) & (df['10-Year Return(%)'] < 20)]
modRisk = modRisk.head(5)
modRisk
```

**Calculating Low Risk**

```python
lowRisk = df[(df['Market Cap(Cr)'] >= 17000) & 
           (df['10-Year Return(%)'] >= 20)]
lowRisk = lowRisk.head(5)
lowRisk
```
